```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def roll_die(num_sides=6, num_rolls=1, show_val=False):
    results = []
    
    # Roll die num_rolls times
    for i in range(num_rolls + 1):
        roll = random.randrange(1, num_sides + 1)
        results.append(roll)
        
        if show_val: 
            print(f"You rolled a {roll}!")
        
    # Plot the rolls in a histogram
    plt.hist(results)
    plt.show()
    
    # Show average value
    print(f"Average value: {np.mean(results)}")
```

```python
roll_die(num_sides=20, num_rolls=1000000)
```

The histogram is surprisingly flat, given that the number of rolls is incredibly high. With fewer rolls, we would be able to understand more of what happened, but at this scale, it is too information dense to make anything of it.

```python
def roll_die_nm(n, m):
    results = []
    for i in range(m + 1):
        roll = random.randrange(1, n + 1)
        results.append(roll)
        
    maxval = np.max(results)
    return maxval
```

```python
vals =  []
for i in range(1000001):
    vals.append(roll_die_nm(20, 2))
    
plt.hist(vals)
plt.show()
print(f"Average value: {np.mean(vals)}")
```

## Recap

The two histograms looks vastly different. The first was incredibly flat, given that there was a roughly equal chance of every side being rolled over 1,000,000 rolls.

The second histogram shows a different story, where the majority of rolls were closer to a natural 20 and much fewer below 10, averaging around 15-16. This makes sense, given that we were recording only the maximum value from 2 rolls of a D-20 die, so it was more likely than not that the values would skew higher.
