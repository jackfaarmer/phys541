```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def fitfunc(x, a, b, c, d, e):
    return a + (b * np.array(x)) + (c * np.sin(e * np.array(x))) + (d * np.cos(e * np.array(x)))

a0 = 0
a1 = 1
a2 = 1
a3 = 1
a4 = 0

rand_nums = np.random.uniform(20, 35, 100)
func = fitfunc(rand_nums, a0, a1, a2, a3, a4) + np.random.normal(0, 0.5, 100)

plt.plot(rand_nums, fitfunc(rand_nums, a0, a1, a2, a3, a4), 'o')
```

```python
def fitfunc_2(x, a, b, c, d, e):
    return a + (b * np.array(x)) + (c * np.sin((d * np.array(x)) + e))

popt_1, pcov_1 = curve_fit(fitfunc, rand_nums, func)
popt_2, pcov_2 = curve_fit(fitfunc_2, rand_nums, func)

plt.plot(rand_nums, func, 'o')
plt.plot(rand_nums, fitfunc(rand_nums, *popt_1), 'b-')
plt.plot(rand_nums, fitfunc_2(rand_nums, *popt_2), 'o-')
```
